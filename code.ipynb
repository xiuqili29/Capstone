```python
d1 = d1.fillna(d1.mean())

d1 = d1.dropna()
```

```python
d1['m1_chn1'] = (d1['m1_chn'] / d1['CNYUSD']) / 1000000000
print((d1))
df = d1
```

```python
df['cpi_d'] = df['cpi_us'] - df['cpi_chn']
df['iip_d'] = df['iip_us'] - df['iip_chn']
df['irr_d'] = df['irr_us'] - df['irr_chn']
df['m1_d'] = df['m1_us'] - df['m1_chn1']
df['rsrv_d'] = df['tr_us'] - df['tr_chn']
df['stock_d'] = df['stock_us'] - df['stock_chn']
df['trade_d'] = df['nt_us'] - df['nt_chn']
```

```python
import datetime as dt

df['month'] = pd.DatetimeIndex(df['DATE']).month
df['year'] = pd.DatetimeIndex(df['DATE']).year
df['Date'] = pd.to_datetime(df['DATE'])
```

```python
y = df['CNYUSD']

plt.plot(df['Date'], y)
plt.title('Foreign Exchange Rate (CNYUSD)')
#plt.show()
```

```python
for params in ['cpi_', 'iip_', 'irr_', 'm1_', 'tr_', 'stock_', 'nt_']:
        
    if params in ['reserves_', 'trade_']:
        plt.plot(df['Date'], df[params+'chn']/1000000000, label='China')
        plt.plot(df['Date'], df[params+'us']/1000000000, label='USA')
    else:
        plt.plot(df['Date'], df[params+'chn'], label='China')
        plt.plot(df['Date'], df[params+'us'], label='USA')
        
    if params == 'cpi_':
        plt.title('Consumer Price Index (CPI)')
    elif params == 'iip_':
        plt.title('Index of Industrial Production (IIP)')
    elif params == 'irr_':
        plt.title('Interest Rates (%)')
    elif params == 'm1_':
        plt.title('Money Supply ($ Billion)')
    elif params == 'tr_':
        plt.title('Reserves ($ Billion)')
    elif params == 'stock_':
        plt.title('Stock Index (Scaled)')
    elif params == 'nt_':
        plt.title('Net Exports ($ Billion)')
    plt.legend()
    # plt.savefig(params + '.png')
    plt.show()
```
